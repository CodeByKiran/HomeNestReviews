```python
import pandas as pd

df = pd.read_csv('../data/processed/reviews_with_summaries.csv')
df.head()
```

```python
overall_summary = " ".join(df['abstractive_summary'].dropna())

print(overall_summary[:1000])
```

```python
positive_reviews = df[df['sentiment'] == 'Positive']

positive_summary = " ".join(positive_reviews['abstractive_summary'].dropna())

print(positive_summary[:1000])
```

```python
negative_reviews = df[df['sentiment'] == 'Negative']

negative_summary = " ".join(negative_reviews['abstractive_summary'].dropna())

print(negative_summary[:1000])
```

```python
neutral_reviews = df[df['sentiment'] == 'Neutral']

neutral_summary = " ".join(neutral_reviews['abstractive_summary'].dropna())

print(neutral_summary[:1000])
```

```python
from collections import Counter

all_summaries = " ".join(df['abstractive_summary'].dropna())

word_counts = Counter(all_summaries.split())

word_counts.most_common(20)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(df['abstractive_summary'].dropna())

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud)
plt.axis("off")
plt.title("Customer Feedback Word Cloud")
plt.show()
```
