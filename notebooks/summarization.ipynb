```python
from tqdm import tqdm
import pandas as pd

import os 
import sys 
sys.path.append(os.path.abspath(".."))

from  src.summarizer import extractive_summary
from  src.summarizer import abstractive_summary

df = pd.read_csv("../data/processed/reviews_with_summaries.csv")

#extractive 
df['extractive_summary'] = df['review_text'].progress_apply(lambda x: extractive_summary(x, num_sentences=1))

#Add abstractive 
df['abstractive_summary'] = df['review_text'].progress_apply(lambda x: abstractive_summary(x))
```
