Comparing NLTK and SpaCy 

```python
import pandas as pd
import nltk
import spacy
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
nltk.download('punkt_tab')

nlp = spacy.load("en_core_web_sm")

df = pd.read_csv("../data/raw/homenest_reviews.csv")

samples = df["review_text"].sample(5, random_state=42)

for text in samples:
    print("RAW:", text)
    print("NLTK:", word_tokenize(text))
    print("spaCy:", [token.text for token in nlp(text)])
    print("-" * 80)
```

Stemming vs lemmatization


```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import spacy

ps = PorterStemmer()
nlp = spacy.load("en_core_web_sm")

text = "The products were running better than expected and deliveries were delayed."

tokens = word_tokenize(text)

stemmed = [ps.stem(t) for t in tokens]

doc = nlp(text)
lemmatized = [token.lemma_ for token in doc]

print("Original:", tokens)
print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
```

StopWord Removal -- To check whether we are loosing any sentiment or not 

```python
from nltk.corpus import stopwords
nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

text = "This product is not good at all"
tokens = word_tokenize(text)

filtered = [t for t in tokens if t.lower() not in stop_words]

print("Original:", tokens)
print("After Stopword Removal:", filtered)
```

POS filtering -->Instead of keeping all words, we only keep Nouns , verbs ,adverbs , adj any Combination

```python
doc = nlp("The product is extremely durable but slightly overpriced.")

filtered = [token.lemma_ for token in doc if token.pos_ in ["NOUN", "ADJ"]]

print(filtered)
```
